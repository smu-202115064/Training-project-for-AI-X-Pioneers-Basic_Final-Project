```python
from glob import glob
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

from python import base


DATA_DIR = base.BASE_DIR / 'data/KOSIS/data'
FIGURES_DIR = base.BASE_DIR / 'figures'


DATA_FILES = [*map(Path, glob(str(DATA_DIR / '*.csv')))]

for file in DATA_FILES:
    print(file.name)
```

```python
def load_data() -> pd.DataFrame:
    """단위: %"""
    df = pd.read_csv(DATA_DIR / '문화예술행사_직접_관람횟수__대중음악_연예_20241129080635.csv', header=None)
    df = df.transpose()
    df.columns = df.iloc[1]
    df = df.drop(df.columns[-11:-1], axis=1)
    df = df[1:]
    # get columns
    columns = df.iloc[0].tolist()
    df = df[1:]
    # set columns
    columns[0] = '연도'
    columns[1] = '관람횟수'
    df.columns = columns
    return df

load_data()
```
